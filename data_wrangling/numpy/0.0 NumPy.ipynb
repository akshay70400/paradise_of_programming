## copy()
-if change in one, No change in other

```python
v=np.array([1,2,3])
c=np.copy(v)
c
```

```python
c[0]=99
print("c", c)
print("v", v)
```

## view()
-creates a reference of existing array

-if change in one, there is change in other

```python
v=np.array([1,2,3])
c=v.view()
c
```

```python
c[0]=99
print("c", c)
print("v", v)
```
