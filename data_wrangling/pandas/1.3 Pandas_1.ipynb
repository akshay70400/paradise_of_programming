## What is Pandas:
-Open source data analysis library written in python
-It leverages the speed and power of Numpy(which is written in C lang) to make datat analysis and preprocessing easy for data scientists
-It provides rich and highly robus data operations

#### Pandas has 2 types of data structures:
SERIES: Its a 1D array with indexes it stores single column or row of data in a dataframe. Capable of holding any one type of data.

DATAFRAMES: Its a tabular spreadsheet like structure representing rows which contains one or more columns. Capable of holding columns pf potentially different types of data.
