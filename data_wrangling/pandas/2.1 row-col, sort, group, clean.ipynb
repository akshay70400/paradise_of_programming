# 9. Cleaning Data: Casting datatypes and Handling missing data

```python
import pandas as pd
import numpy as np

dfp=pd.read_csv('C:\\Users\\admin\\Desktop\\Data_Science\\Data_for_ananlysis\\2019 developer survey Stackoverflow\\survey_results_public.csv'\
                     ,index_col='Respondent')

dfs=pd.read_csv('C:\\Users\\admin\\Desktop\\Data_Science\\Data_for_ananlysis\\2019 developer survey Stackoverflow\\survey_results_schema.csv'\
                     ,index_col='Column')

dictionary={
    "first":["Corey","Jane","John","Chris",np.nan,None,'NA'],
    "last":["Schafer","Doe","Doe","Schafer",np.nan,np.nan,'Missing'],
    "email":["CS@gmail","JaD@yahoo","JoD@facebook","ChS@gmail",None,np.nan,"anonymous@gmail"],
    "age":['33','33','63','36',None,None,'Missing']
}

ddf=pd.DataFrame(dictionary)
```
