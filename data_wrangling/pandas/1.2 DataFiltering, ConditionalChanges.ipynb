# Conditional changes 

```python
details = {
    'Name' : ['Ankit', 'Aishwarya','Ankit', 'Aishwarya', 'Shaurya', 'Shivangi'],
    'Maths' : [80, 40, 50, 70,42, 61],
   'Sci' : [83, 21, 43, 61,72, 71],
    'Hindi' : [53, 21, 73, 11,92, 11],
    'Chem' : [73, 41, 43, 41,82, 91],
    'Bio' : [93, 41, 23, 41,45, 51],
    'Sanskrit' : [73, 81, 53, 41,72, 21],
}
import pandas as pd
d1=pd.DataFrame(details)

d1
```

```python
d1["total"]=0
```

```python
d1["total"]=d1["Maths"]+d1["Sci"]+d1["Hindi"]+d1["Chem"]+d1["Bio"]+d1["Sanskrit"]
d1
```

```python
d1["%"]=0
d1
```

```python
d1["%"]=(d1["total"]/600)*100
d1
```

### Example: Grades

```python
d1.loc[d1["%"]<45]
```

```python
d1.loc[(d1["%"]>45)&(d1["%"]<60),["Grade"]]="PASS"
```

```python
d1.loc[(d1["%"]>60)&(d1["%"]<70),["Grade"]]="FIRST CLASS"
```

```python
d1.loc[(d1["%"]>70)&(d1["%"]<100),["Grade"]]="DISTINCTION"
```

```python
d1.loc[d1["%"]<45,["Grade"]]="FAIL"
```

```python
>>> df = pd.DataFrame([[1, 2], [4, 5], [7, 8]],
...      index=['cobra', 'viper', 'sidewinder'],
...      columns=['max_speed', 'shield'])
```

```python
df
```

```python
#Single label. Note this returns the row as a Series.

df.loc['viper']
```

```python
#List of labels. Note using ``[[]]`` returns a DataFrame.
df.loc[['viper', 'sidewinder']]
```

```python
#Single label for row and column
df.loc['cobra', 'shield']
```

```python
#Slice with labels for row and single label for column. As mentioned
#above, note that both the start and stop of the slice are included.

df.loc['cobra':'viper', 'max_speed']
```

```python
#Boolean list with the same length as the row axis

df.loc[[False, False, True]]
```

```python
#Alignable boolean Series:
df.loc[pd.Series([False, True, False],index=['viper', 'sidewinder', 'cobra'])]
```

```python
#Index (same behavior as ``df.reindex``)
df.loc[pd.Index(["cobra", "viper"], name="foo")]
```

```python
#Conditional that returns a boolean Series
df.loc[df['shield'] > 6]
```

```python
#Conditional that returns a boolean Series with column labels specified
df.loc[df['shield'] > 6, ['max_speed']]
```

```python
#Callable that returns a boolean Series
df.loc[lambda df: df['shield'] == 8]
```

```python
#Set value for all items matching the list of labels
df.loc[['viper', 'sidewinder'], ['shield']] = 50
df
```

```python
#Set value for an entire row

df.loc['cobra'] = 10
df
```

```python
#Set value for an entire column

df.loc[:, 'max_speed'] = 30
df
```

```python
#Set value for rows matching callable condition
df.loc[df['shield'] > 35] = 0
df
```
