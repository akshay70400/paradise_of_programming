# FILTERING

```python
people={
    "first":["Corey","Jane","John"],
    "last":["Schafer","Doe","Doe"],
    "email":["CS@gmail","JaD@gmail","JoD@gmail"]
}
df_people=pd.DataFrame(people)
df_people
```

```python
df_people["last"]=='Doe'
```

```python
df_people[df_people["last"]=='Doe']
```

```python
filt = (df_people["last"]=='Doe')
print(filt)
```

```python
df_people[filt]
```

```python
df_people.loc[filt]
#loc can accept boolean series
```

```python
df_people[filt]['email']
```

```python
df_people.loc[filt,'email']
```

```python
# & : and
# | : or
```

```python
filt=(df_people["last"]=="Doe") & (df_people["first"]=="John")
df_people.loc[filt]
```

```python
filt=(df_people["last"]=="Schafer") | (df_people["first"]=="John")
df_people.loc[filt]
```

```python
#TO get opposite of filter
filt=(df_people["last"]=="Schafer") | (df_people["first"]=="John")
df_people.loc[-filt]
```

# MODYFYING DATA WITHIN DATAFRAME

```python
people={
    "first":["Corey","Jane","John"],
    "last":["Schafer","Doe","Doe"],
    "email":["CS@gmail","JaD@gmail","JoD@gmail"]
}
df_people=pd.DataFrame(people)
df_people
```

```python
df_people.columns
```

```python
df_people.columns = ["First_name","Last_name","EMails"]
df_people
```

```python
df_people.columns=[x.upper() for x in df_people.columns]
df_people
```

```python
df_people.columns=df_people.columns.str.replace(" ","_")
df_people
```

```python
df_people.columns=[x.lower() for x in df_people.columns]
df_people
```

```python
df_people.rename(columns={'first_name':'first','last_name':'last'})
#This doesnt change original df
```

```python
df_people
```

```python
df_people.loc[2]=["John","Smith","JS@gmail"]
df_people
```

```python
df_people.loc[2,["last_name",'emails']]=["Doe","JoD@gmail"]
df_people
```

```python
filt=(df_people["emails"]=="JoD@gmail")
df_people[filt]["last_name"]='Smith'

#THis gives error
```

```python
filt=(df_people["emails"]=="JoD@gmail")
df_people.loc[filt,"last_name"]='Smith'
df_people
```

```python
df_people["emails"]=df_people["emails"].str.lower()
df_people
```
