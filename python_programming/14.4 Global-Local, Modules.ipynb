### Variable: Global, Local, Nonlocal

```python
a=100
def my():
    global a #This is a way to change a global variable inside a function
    a=a*2
    print("a inside function",a)
    
my()
print("a outside function",a)
```

```python
#LOCAl variable
def fun():
    b=10
    print("b inside function ",b)

fun()

#print(b) this will give error
```

```python
x="this is global"
print(x)
def new():
    global x
    y="local"
    x=x+"@@@@"+x
    print(x)
    print(y)
    
new()
print("changed x: ",x)
#print(y) this will give an erro as y in not defined
```

#### Non Local
-Nonlocal variable are used in nested function whose local scope is not defined. This means, the variable can be neither in the local nor the global scope.

-We use nonlocal keyword to create nonlocal variable.

```python
x=200 #Global weight outside function
def outer():
    #global x
    x=10 #Local weight inside function_weight
    
    def nested():
        global x
        x="nonlocal" #Non local weight function dryfruits
        print("inner x: ",x)
        
    nested()
    print("outer x: ",x)
    
outer()
print("Exterme Out: ",x)
```
