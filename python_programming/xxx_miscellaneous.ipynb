### List Comprehension

newlist=[x for x in (iterable) if (condition)]

List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

Example:

Based on a list of fruits, you want a new list, containing only the fruits with the letter "a" in the name.

Without list comprehension you will have to write a for statement with a conditional test inside:

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)
```

```python
#The expression is the current item in the iteration, but it is also the outcome, which you can manipulate before it ends up like a list item in the new list:
newlist1 = [x.upper() for x in fruits]
newlist1
```

```python
#You can set the outcome to whatever you like:
newlist2 = ['hello' for x in fruits]
newlist2
```

```python
#The expression can also contain conditions, not like a filter, but as a way to manipulate the outcome:
newlist3 = [x if x != "banana" else "orange" for x in fruits]
newlist3
#Return the item if it is not banana, if it is banana return orange".
```

### Dictionary Comprehension

```python
keys = ['a','b','c','d','e']
values = [1,2,3,4,5]  

myDict = { k:v for (k,v) in zip(keys, values)}  
```

```python
myDict = {x: x**2 for x in [1,2,3,4,5]}
```

```python
newdict = {x: x**3 for x in range(10) if x**3 % 4 == 0}
```

### any()

The any() function is an inbuilt function in Python which returns true if any of the elements of a given iterable( List, Dictionary, Tuple, set, etc) are True else it returns False. 0

iterables that can be passed: list, tuple, set, dict, string, fset
returns false for: 0, False, None, empty, empty string

```python
# All elements of list are true
l = [ 4, 5, 1]
print(any( l ))
  
# All elements of list are false
l = [ 0, 0, False]
print(any( l ))
  
# Some elements of list are
# true while others are false
l = [ 1, 0, 6, 7, False]
print(any( l ))
  
# Empty List
l = []
print(any( l ))
```
