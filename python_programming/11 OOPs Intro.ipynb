#### 4 principles of OOPs:
    
    Abstraction - Hide implementation details and only show essential details.
    
    Encpsulation -  Wrapping/packing the data and functions in a single entity. (public, private, protected)
    
    Inheritance - taking the methods from the parent class
    
    Polymorphism - multiple forms of a method
        Ex. + sign used for arithmetic operation(5+6) as well as fro string concatenation("Hi"+"Akshay")

### Abstraction: (Hide implementation details and show essential details)

-Abstract Class - A class which has abstract method.

-Abstract Method - A method which has declaration and not definition/implementation

-Concrete class - A class which does not have any abstract method.

-Object cant instantiate or access the abstract class.

-Need to import from module abc (abstract base class).

```python
from abc import ABC, abstractmethod

class Velocity_class(ABC): #Abstract class
    
    @abstractmethod #Decorator
    def mock(self):
        print("Daily mock")
        
    def interview(self):
        print("On sunday")
        
# obj=Velocity_class()
# """
# ERROR: Can't instantiate abstract class Velocity_class with abstract methods mock
# """

class xyz(Velocity_class): #Concret class
    def mock(self):
        print("xyz mock is working")
    
x=xyz()
x.mock()
x.interview()
```
