#### partition()
-converts string into tupple.

-takes exactly one arguement

-same as split() BUT 

    returns tupple instead of list in split()
    
    keeps the separator element unlike split()
    
-@the separator arguement is always kept in middle in tupple

```python
#full_name="Ajay Annaji Kumbhar   @   from Pune "
#full_name.split()


#--> ['Ajay', 'Annaji', 'Kumbhar', '@', 'from', 'Pune']
```

```python
example=("Python is easy to learn")

print(example)
example=example.partition("is")
print(example)
```

####  join()
-Against of split()

-"DHONI"split() --> [D,H,O,N,I]

-[D,H,O,N,I] join() --> DHONI

-it joins iterables into a string

-Concatenate any number of strings.

-Syntax:   separator.join(iterable, /)

-Example:   '.'.join(['ab', 'pq', 'rs']) -> 'ab.pq.rs'

```python
'...'.join(['ab', 'pq', 'rs'])
#applicable to tupple also
```

```python
words=["India", "is", "best"]
print(words)
".".join(words)
```

```python
"'".join(words)
```

```python
"".join(words)
```

```python
" ".join(words)
```

```python
words=["India", "is", "best"]
print(words)

separator="''"
separator.join(words)
```

```python
words=["India", "is", "best"]
print(words)

separator1="#"
separator1.join(words)
```

####  replace()
to replace a string by another

original string will be returned if substring is not found

```python
indian_players="Virat is captain of India"

indian_players.replace("Virat","Dhoni")
```

```python
print(indian_players) #Notice no change in original variable as string is immutable
```

```python
indian_players="Virat is captain of India"

indian_players.replace("Yuvraj","Dhoni")
```

```python
ipl_winner="KKR MI MI RCB RCB CSK HYD RAJ RCB KKR MI CSK"
print(ipl_winner)

#Replace ALL
ipl=ipl_winner.replace("MI","Pune")
print(ipl)

#Replace first two
ipl2=ipl_winner.replace("MI","Pune",2)
print(ipl2)
```

####  ljust()
sample.ljust(width, fillchar=' ', /)

Return a left-justified string of length width

```python
sample="python"
print(sample)
sample=sample.ljust(10,"+")
print(sample)
```
