# String Methods


#### isprintable()
returns True if the string is printable

Printable characters :
Letter and symbols,
Digits,
Whitespaces,
Punctuation.

Non Printable : \n \t

```python
my_address="Pune-411     019 . , [], @"
my_address.isprintable()
```

```python
my_address2="Satara\n-411019"
my_address2.isprintable()
```

```python
my_address3="Satara\t-411019"
my_address3.isprintable()
```

####  isspace()

```python
telescope=" "
telescope.isspace()
```

```python
telescope1="\t"
telescope1.isspace()
```

```python
telescope2=""
telescope2.isspace()
```

```python
telescope3="     a   kk  l"
telescope3.isspace()
```

```python
telescope4="\n"
telescope4.isspace()
```

#### split()
this is VIMP method

syntax:   .split(sep=None, maxsplit=-1)


it converts a string into the list


it breaks up the string at specific separator and returns a list of string


By default separator is space. {-1 (the default value) means no limit.}

    .split() --> Takes space as separator

    .split(" ") --> Takes space as separator
    
    .split("") --> Error
    
Return a list of the words in the string, using sep as the delimiter string.

The element inside () gets eliminated from list

Maxsplit:
    
    Maximum number of splits to do.

    -1 (the default value) means no limit.

```python
full_name="Ajay Annaji Kumbhar   @   from Pune "
print(full_name)
full_name.split()
```

```python
my_grocery="MilkEgg.Bread"
print(my_grocery)

print(my_grocery.split())
print(my_grocery.split("."))
print(my_grocery.split("g"))
print(my_grocery.split("a"))

k=my_grocery.split(".")
print(type(k))
```

```python
food=" roti,    bread,rice, paneer,chicken"

a=food.split()
print(a)

b=food.split(",")
print(b)

c=food.split("p")
print(c)

#d=food.split("") --> This returns an error as Empty Separator
print("\n")

#MAX SPLIT returns maximum number of splits you want (n+1)
print("0 :",food.split(",",0))
print("1 :",food.split(",",1))
print("2 :",food.split(",",2))
print("3 :",food.split(",",3))
print("4 :",food.split(",",4))
print("5 :",food.split(",",5)) # 5 retuns same as 4

print("with 'r' :",food.rsplit(",",2))
```

####  splitlines()
-Return a list of the lines in the string, breaking at line boundaries.

-.splitlines(keepends=False/True)
    
    False is default

-Line breaks are not included in the resulting list unless keepends is given and
true.

```python
optionals="anthro,\nsocio\npol,\ngeo\nhis"
print(optionals)

optionals.splitlines()
```

```python
optionals.splitlines(keepends=True)
```

```python
optionals.splitlines(keepends=False)
```

#### lstripe()
-Return a copy of the string with LEADING whitespace removed.

-If chars is given and not None, remove characters in chars instead.

```python
random="        this    is  a lie"
print(random)
print(len(random))

#random_new=random.lstrip() ->Initial spaces by default 

random_new=random.lstrip("hi s t") 
#alphabets can be put once
#alphabets not necessary to be in order
print(random_new)
print(len(random_new))
```

```python
website="htttttttppp://www.facebook.com"
print(website)
website_new1=website.lstrip("htp://")
print(website_new1)
```

```python
website="https://www.facebook.com"
#stripes Only the leading characters
website_new1=website.lstrip(".com")
print(website_new1)
```

####  rstrip()

```python
websites="https://www.facebook.   coooommmm   "
print(websites)
website_new=websites.rstrip(".com ")
print(website_new)
```
