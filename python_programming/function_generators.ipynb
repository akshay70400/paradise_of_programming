# Generators

has advantage over lists
it does not hold the result/(all elements) in the memory at once rather it returns the result one by one only when asked
it makes code more readable


```python
from typing import List
def square(nums: List) -> List:
    result = []
    for i in nums:
        result.append(i*i)
    return result
my = square([1,2,3,4])
print(my)
```

##### we can get rid of having to create an empty list and then appending it using generators(yield)

```python
from typing import List
def square(nums: List) -> List:
    for i in nums:
        yield (i*i) 
sqr_nums = square([1,2,3,4])
print(sqr_nums)
```

```python
from typing import List
def square(nums: List) -> List:
    for i in nums:
        yield (i*i) 
sqr_nums = square([1,2,3,4])
print(next(sqr_nums))
print(next(sqr_nums))
```

```python
# Stop iteration exception Error : if the generators runs out of list/elements
# so you can use for loop over generators
for i in sqr_nums:
    print(i)
```

```python
# also list comprehension can be used
my = [i*i for i in [1,2,3,4]]
my
```

```python
my = (i*i for i in [1,2,3,4])
print(my)
print(next(my))
```
