#### EXERCISE: Divisible by 2,3,5

```python
n=int(input("Enter a number : "))
if n%2==0:
    if n%3==0:
        if n%5==0:
            print("by 2,3 and 5")
        else:
            print("by 2 and 3")
    elif n%5==0:
        if n%3==0:
            print("by 2,3 and 5")
        else:
            print("by 2 and 5")
    else:
        print("by 2")    
elif n%3==0:
    if n%5==0:
        print("by 5 and 3")
    else:
        print("by 3")
elif n%5==0:
    print("by 5")
else:
    print("Number is not divisible by 2,3 or 5.")
```

#### EXERCISE: Greatest of 3 numbers

```python
n1=float(input("Enter first number: "))
n2=float(input("Enter second number: "))
n3=float(input("Enter third number: "))

if n1>=n2 and n1>=n3:
    print(n1," is greatest")
elif n2>=n1 and n2>=n3:
    print(n2," is greatest")
elif n3>=n1 and n3>=n2:
    print(n3," is greatest")
```
